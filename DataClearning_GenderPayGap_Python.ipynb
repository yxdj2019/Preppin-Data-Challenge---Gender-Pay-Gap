```python
import pandas as pd
import numpy as np
import os
```

# Return a list containing the names of the files in the path and read files; Add Year and Report columns

```python
filename_list=[]
df_list=[]
for filename in os.listdir('Inputs'):
    if filename.endswith('.csv'):
        filename_list.append(filename[:-4].split('- ')[1])
        df=pd.read_csv('Inputs/'+filename)
        # add year
        df.insert(0,'Year',filename[:-4].split('- ')[1][0:4])  # add 'Year'
        df.insert(1,'Report',filename[:-4].split('- ')[1])     # add 'Report'
        df_list.append(df)
# df2=pd.read_csv('Inputs/UK Gender Pay Gap Data - 2018 to 2019.csv')
# df3=pd.read_csv('Inputs/UK Gender Pay Gap Data - 2019 to 2020.csv')
# df4=pd.read_csv('Inputs/UK Gender Pay Gap Data - 2020 to 2021.csv')
# df5=pd.read_csv('Inputs/UK Gender Pay Gap Data - 2021 to 2022.csv')
```

# Union all the data and Clearning

```python
df_all=pd.concat(df_list,ignore_index=True)
```

```python
# check if has duplicates  ---- no
df_all.duplicated().isna().any()
```

```python
# keep necessary columns
df_all_new=df_all.loc[:,['Year','Report','EmployerName', 'CurrentName', 'EmployerId', 'EmployerSize', 'DiffMedianHourlyPercent']]
```

```python
df_all_new.head()
```

```python
df_all_new=df_all_new.convert_dtypes()
```

```python
# update EmployerName based on the name in the latest year
for id in df_all_new['EmployerId'].unique():
    df_all_new.loc[df_all_new['EmployerId']==id,'EmployerName']=df_all_new[(df_all_new['EmployerId']==id)&(df_all_new['Year']==df_all_new[df_all_new['EmployerId']==id].Year.max())].CurrentName.values[0]

# Notice that we use the '.values[0]' to extract the value in the seriers; If not using it, results will be not correct
```

```python
# drop 'CurrentName'
df_all_new.drop(columns='CurrentName', inplace=True)
```

```python
df_all_new.head(10)
```

```python
# add 'Pay Gap'

# for equal pay
df_all_new.loc[df_all_new['DiffMedianHourlyPercent']==0,'Pay Gap']="In this organisation, men's and women's median hourly pay is equal."
```

```python
# for less pay
df_all_new.loc[df_all_new['DiffMedianHourlyPercent']>0,'Pay Gap']="In this organisation, women's median hourly pay is X% lower than men's."
df_t=df_all_new[df_all_new['DiffMedianHourlyPercent']>0]
# use apply function
df_all_new.loc[df_all_new['DiffMedianHourlyPercent']>0,'Pay Gap']=df_t.apply(lambda x: x['Pay Gap'].replace('X',str(x['DiffMedianHourlyPercent'])),axis=1)

# apply() function calls the lambda function and applies it to every row or column of the dataframe and 
# returns a modified copy of the dataframe
# --------------------
## Notice: axis =1 说明apply的func以DataFrame的每一行为单位工作, 即参数x代表DataFrame的每一行; 
## 那么x['Pay Gap']代表一个元素即string,所以只能用.replace而不能再用str.replace
```

```python
# for more pay
df_all_new.loc[df_all_new['DiffMedianHourlyPercent']<0,'Pay Gap']="In this organisation, women's median hourly pay is X% higher than men's."
df_t1=df_all_new[df_all_new['DiffMedianHourlyPercent']<0]
df_all_new.loc[df_all_new['DiffMedianHourlyPercent']<0,'Pay Gap']=df_t1.apply(lambda x: x['Pay Gap'].replace('X',str(abs(x['DiffMedianHourlyPercent']))), axis=1)
```
